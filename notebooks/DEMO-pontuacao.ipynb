## Como são os currículos?

```python
# Como são os XMLs
XML_logado = '../../data/raw/CVs/8090714349274051.xml'
XML_nao_logado = '../../data/raw/CVs/curriculo.xml'
file = XML_logado
tree = etree.parse(file)
root = tree.getroot()
CPF = root.find('DADOS-GERAIS').attrib['CPF']
NOME = root.find('DADOS-GERAIS').attrib['NOME-COMPLETO']
ID = root.attrib['NUMERO-IDENTIFICADOR']
[CPF, ID, NOME]
root.getchildren()
from lxml.etree import ParseError
from lxml.etree import ParserError
file = XML_nao_logado
tree = etree.parse(file)
root = tree.getroot()
try:
    CPF = root.find('DADOS-GERAIS').attrib['CPF']
except (ParserError, ParseError, KeyError) as error:
    CPF = ''
    print('CPF indefinido')
NOME = root.find('DADOS-GERAIS').attrib['NOME-COMPLETO']
ID = root.attrib['NUMERO-IDENTIFICADOR']
[CPF, ID, NOME]
root.getchildren()
```
